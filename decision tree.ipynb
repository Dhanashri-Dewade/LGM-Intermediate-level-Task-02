## Spliting the data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(x,y,test_size = 0.25, random_state= 0)
 
```

## Now let us define the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion='gini',min_samples_split=2)
dtree.fit(x,y)
print('--------Decision Tree Classifer Created-------')
```

```python
y_pred=dtree.predict(x_test)
```

## Visualizing the Decision tree

```python
from sklearn import tree
names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
classes=['Iris-setosa','Iris-versicolor','Iris-virginica']
fig,axex=plt.subplots(nrows=1,figsize=(4,4),dpi=300)
tree.plot_tree(dtree,feature_names=names,class_names=classes,filled=True);
```
